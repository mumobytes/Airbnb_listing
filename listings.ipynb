```python
import numpy as np
import csv
import matplotlib.pyplot as pyt
import pandas as pd 
import seaborn as sns 
```

```python
path = r"C:\Users\user\.vscode-py\data\Airbnb+Data\Airbnb Data\Listings.csv"
listing_df = pd.read_csv(path, encoding="ISO-8859-1 ", low_memory =False)
listing_df
```

```python
listing_df.describe()
listing_df.columns
```

```python
listing_df.info()
```

```python
listing_df['host_since']= pd.to_datetime(listing_df['host_since'])
```

```python
listing_df.info()
```

```python
data = listing_df[listing_df['city']=='Paris'].loc[:, ["host_since","neighbourhood","city","accommodates","price"]]
                
data.info()
```

```python
paris_listings = data
paris_listings
```

```python
paris_listings.info()
```

```python
paris_listings.isna().sum()
```

```python
paris_listings_neighbourhood = (
    paris_listings
    .groupby("neighbourhood")
    .agg({"price":"mean"})
    .sort_values("price")
)
paris_listings_neighbourhood

```

```python
paris_listings_accommodates = (
    paris_listings
    .query ("neighbourhood == 'Elysee'")
    .groupby("accommodates")
    .agg({"price" : "mean"})
    .sort_values("price")
)
paris_listings_accommodates
```

```python
paris_listings_over_time= (
    paris_listings
    .set_index("host_since")
    .resample('YE')
    .agg({
        "neighbourhood":"count",
        "price":"mean"
    })
)
paris_listings_over_time
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
plt.barh( paris_listings_neighbourhood['neighbourhood'], paris_listings_neighbourhood['price'], color = "m")
plt.title("Average Price of Airbnb in Paris")
plt.ylabel("Neighbourhood in paris")
plt.xlabel("Price per night (Euros)")
sns.despine()
plt.show()
```
